## Mapa do resultado do 2º turno das Eleições Presidenciais de 2018 por município

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# O arquivo SHP da malha territorial municipal pode ser encontrado no site do IBGE ou através do link https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2020/Brasil/BR/BR_Municipios_2020.zip

# Carregar o shapefile
df_mapa = gpd.read_file('BR_Municipios_2020.shp')
```

```python
# Exibir o mapa
df_mapa.plot();
```

```python
# Expandir o espaço para desenhar o mapa
fig, ax = plt.subplots(1, figsize=(12, 12))

# Desenhar o mapa
df_mapa.plot(ax=ax)

# Retirar os eixos
ax.set_axis_off()
plt.show();
```

```python
# Carregar o arquivo com o resultado da eleição dividido por município/zona
df_votacao = pd.read_csv('votacao_candidato_munzona_2018_BR.csv', sep=';')
df_votacao.head().T
```

```python
# Separar os dados, pois precisamos apenas os dados do 2º turno
df_votacao = df_votacao[df_votacao['NR_TURNO'] == 2]
df_votacao.sample(3).T
```

```python
# Verificar quantos municípios únicos existem no conjunto de dados e pegar por código de município
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Retornou 5708, mas o Brasil possui 5570 municípios. Verificar as cidades fora do Brasil que tiveram eleitores
df_votacao[df_votacao['SG_UF'] == 'ZZ'].sample(4).T
```

```python
# Quantidade de cidades fora do Brasil e que estão no conjunto de dados
df_votacao[df_votacao['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
# Eliminar as cidades que computaram votos de eleitores de fora do Brasil
df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
```

```python
# Verificar a quantidade de municípios
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Os dados estão separado por município/zona
# Municípios pequenos normalmente têm apenas uma zona eleitoral, mas municípios maiores podem ter várias zonas eleitorais

# Verificando quantas zonas eleitorais tem Recife, por exemplo
df_votacao[df_votacao['NM_MUNICIPIO'] == 'RECIFE'].T
```

```python
# A tabela acima mostra que Recife possui 11 zonas eleitorais, pois cada zona eleitoral tem duas linhas, sendo uma para cada candidato
```

```python
# Eliminar colunas do dataframe que não serão utilizadas
df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]
df_votacao.sample(5)
```

```python
# Agregar os dados por município e candidato
# Cada município deve ter apenas duas linhas: a quantidade de votos do candidato 13 e outra do candidato 17
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# Fazer o groupby, reindexar o índice e salvar o dataframe
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()
df_votacao.sample(5)
```

```python
# Verificar novamente a cidade do Recife para confirmar que está tudo certo
df_votacao[df_votacao['NM_MUNICIPIO'] == 'RECIFE']
```

```python
# Ordenar o dataframe pela quantidade de votos nominais e deixar apenas o primeiro registo de cada município
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)
```

```python
# Ordenar
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

# Dropar linhas duplicadas, mantendo a primeira
df_votacao.drop_duplicates('CD_MUNICIPIO', keep='first', inplace=True)
df_votacao.head(10)
```

```python
# Verificar novamente a cidade do Recife
df_votacao[df_votacao['NM_MUNICIPIO'] == 'RECIFE']
```

```python
# Criar uma coluna representando o resultado da eleição daquele município, sendo o resultado True se o candidato 17 venceu naquele município, e False se o candidato 13 venceu
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
# Verificar o resultado
df_votacao.head(10)
```

```python
# Mostrando alguns valores do dataframe do geopandas
df_mapa.sample(3)
```

```python
# Tentando encontrar o município no dataframe do TSE pelo código no dataframe do geopandas
df_votacao[df_votacao['CD_MUNICIPIO'] == 3115508]
```

```python
# O porquê de não podermos usar o nome do município para juntar os dados: mais de um município com o mesmo nome
df_votacao[df_votacao['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```python
# Recuperar a tabela de correspondência entre IBGE e TSE direto do GitHub do Estadão
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'
df_equivalencia = pd.read_csv(link)
df_equivalencia.head()
```

```python
# Preparar os dataframes
```

```python
# Reindexar o dataframe de votação pelo código do TSE
df_votacao_novo = df_votacao.copy()
df_votacao_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True) # Torna a coluna 'CD_MUNICIPIO' o index e NÃO exclui ela do dataframe
df_votacao_novo.head()
```

```python
# Reindexar o dataframe de equivalência também pelo código do TSE
df_equivalencia_novo = df_equivalencia.copy()

# Torna a coluna 'COD_TSE' o index e NÃO exclui ela do dataframe
df_equivalencia_novo.set_index('COD_TSE', drop=False, inplace=True) 
df_equivalencia_novo.head()
```

```python
# Agora que os dois dataframes (votação e equivalência) estão indexados pelo mesmo código, podemos criar um novo dataframe
df_votacao_equivalencia = df_votacao_novo.join(df_equivalencia_novo)
df_votacao_equivalencia.head()
```

```python
# Unificar o dataframe do geopandas com o df_votacao_equivalencia
```

```python
df_votacao_equivalencia.info()
```

```python
df_mapa.info()
```

```python
# Transformar a coluna GEOCOD_IBGE do df_votacao_equivalencia
df_votacao_equivalencia['GEOCOD_IBGE'] = df_votacao_equivalencia['GEOCOD_IBGE'].astype(str)
df_votacao_equivalencia.info()
```

```python
# Unificar os índices
# Mudando o índice do df_mapa
df_mapa.set_index('CD_MUN', drop=False, inplace=True)
```

```python
# Mudar o índice do df_votacao_equivalencia
df_votacao_equivalencia.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```python
# Juntar os dataframes
df_mapa_novo = df_mapa.join(df_votacao_equivalencia)
df_mapa_novo.head().T
```

```python
# Desenhando o mapa
# Criar a referência para o código de cor (False/0 = red, True/1 = green)
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['red', 'green'])

# Desenhando o mapa
fig, ax = plt.subplots(1, figsize=(14, 14))
ax = df_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=False, linewidth=0.1, ax=ax, edgecolor='grey')

# Escondendo os eixos
ax.set_axis_off()

# Colocando um título
ax.set_title('Resultado Eleição 2018 por município / 2º Turno', fontdict={'fontsize': '22', 'fontweight': '3'})

# Criando uma nota de rodapé
ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal Estado de São Paulo, 2019',
            xy=(0.1, 0.08), xycoords='figure fraction', 
            horizontalalignment='left', verticalalignment='top', fontsize=10, color='#555555')

# Alterando a legenda
import matplotlib.patches as mpatches
red = mpatches.Patch(color='red', label='Fernando Haddad')
green = mpatches.Patch(color='green', label='Jair Bolsonaro')
plt.legend(handles=[red, green])
plt.show()
```
